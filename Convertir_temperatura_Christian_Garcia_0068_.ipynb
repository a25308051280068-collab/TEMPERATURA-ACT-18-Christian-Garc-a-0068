## 1. Fórmulas de Conversión

Para convertir temperaturas, utilizamos relaciones matemáticas basadas en los puntos de congelación (0 °C / 32 °F) y ebullición (100 °C / 212 °F) del agua.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
* **De Fahrenheit a Celsius:**
    $$°C = (°F - 32) \times \frac{5}{9}$$

---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25 °C a °F (Temperatura ambiente)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$.
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar 32:** $45 + 32 = 77$.
* **Resultado:** 25 °C equivalen a **77 °F**.

### Ejemplo 2: Convertir 98.6 °F a °C (Temperatura corporal)
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$.
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$.
3.  **Restar 32:** $98.6 - 32 = 66.6$.
4.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$.
* **Resultado:** 98.6 °F equivalen a **37 °C**.

## 3. Código en Python

Este programa utiliza funciones específicas y un ciclo interactivo para facilitar el uso al usuario.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú principal y gestiona la lógica"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")
        
        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```

### Explicación técnica del código:
* **`while True`:** Crea un bucle infinito que permite realizar múltiples conversiones sin tener que reiniciar el programa.
* **`float()`:** Se usa para convertir la entrada del usuario (que siempre es texto) a un número decimal, permitiendo temperaturas como "36.5".
* **`f-strings` (ej. `f"{valor:.2f}"`):** Se utilizan para mostrar los resultados de forma limpia, limitando los decimales a dos lugares para mayor claridad visual.
* **`break`:** Es la instrucción que rompe el ciclo cuando el usuario decide seleccionar la opción 3.

````python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú principal y gestiona la lógica"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa. ¡Hasta luego!")
            print("¡Christian Garcia NL = 21!")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
````
